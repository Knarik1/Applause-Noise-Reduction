```
import os
import mir_eval
from IPython.display import Audio, display
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```
%cd /content/drive/My Drive/Applause-Noise-Reduction
os.getcwd()
```

```
song_path = './data/Songs/test_song/Carlos Gonzalez - A Place For Us/mixture16.wav'
sr, song = wavfile.read(song_path)

estimated_song_path = './results/RNN/estimated_song.wav'
sr_est, estim_song = wavfile.read(estimated_song_path)
```

```
# check signal lengths 
print("song length", len(song))
print("estim song length", len(estim_song))
print("trimmed song length", len(song[:len(estim_song)]))
```

```
print("Estimated Song")
display(Audio(estim_song, rate=sr_est))
print("Song")
display(Audio(song, rate=sr))

fig, ax = plt.subplots(figsize=(18,7))
ax.plot(estim_song, label="Estimated Song");
ax.plot(song, label="Song", alpha=0.5);
plt.legend(loc="upper right");
```

```
# Evaluate using mir_eval
scores = mir_eval.separation.evaluate(
    song[:len(estim_song)], estim_song, window=480, hop=241)

# print nicely formatted and aggregated scores
print(scores)
```
